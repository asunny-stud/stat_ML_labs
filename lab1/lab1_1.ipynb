Определить (с обоснованием), зависимы или независимы следующие события:

**(a)** Несовместные события;

$P(AB) = 0$ - пересечение несовместных событий

Для независимых событий $P(AB) = P(A)P(B)$

$P(A)P(B) = 0$

Тогда несовместные события могут быть независимы только при условии, $P(A) = 0$ или $P(B) = 0$

**(b)** События, образующие $\sigma$-алгебру $\Sigma$ в пространстве ($\Omega$,$\Sigma$,$P$);

По условию $\sigma$-алгебры:

1) $\Omega \in \Sigma$
2) $A \in \Sigma$, то $A^C \in \Sigma$
3) $A_1, A_2, ... \in \Sigma$, то $\bigcup\limits_{i=1}^{\infty}A_{i}\in\Sigma$

Рассмотрим тривиальный случай: $\Sigma=\left\lbrace\varnothing,\Omega\right\rbrace$

$P(\varnothing) = 0$ - пустое событие

$P(\Omega) = 1$ - достоверное событие 

Они также являются дополнениями друг друга, а следовательно несовместны, тогда из (а) следует, что они независимы при условии, что одно из них равно 0. $P(\varnothing) = 0$ ==> события тривиальной $\sigma$-алгебры независимы. 



Рассмотрим нетривиальный случай $\sigma$-алгебры и проверим являются ли события, образующие ее, независимыми в совокупности, т.е. $P(\bigcap\limits_{i=1}^{n}) = \prod\limits_{i=1}^{n} P(A_i)$ для любого подмножеста индексов.

Пусть $A \in \Sigma$, $P(A) \in (0,1)$ (нетривиальное событие), тогда $A^C \in \Sigma$, $A^C = 1 - P(A) \in (0, 1)$

$P(A\cdot A^C)=0$, эти события несовместные, следовательно из (а), они будут независимы если одно из них равно $0$.

Тогда для любого нетривиального $A \in \Sigma$ будет $A^C \in \Sigma$, которые являются зависимыми ==> все события, образующие нетривиальную $\sigma$-алгебру не являются независимыми в совокупности. 

**(c)** События, имеющие одинаковую вероятность

$P(A) = P(B)$

Рассмотрим на примере: 

1. Кидаем 1 монетку (А - орёл, В - решка). $P(A) = P(B) = 1/2$. $P(AB) = 0$ - невозможно выпадение орла и решки одновременно. $P(A) \cdot P(B) = 1/4 \neq 0$ - следовательно события зависимы.

2. Кидаем 2 монетки. (А - на 1 монетке решка, В - на 2 монетке решка). $P(A) = P(B) = 1/2$. $P(AB) = 1/4$ - две решки на монетках. $P(A) \cdot P(B) = 1/4$ - следовательно события независимы.

Независимость событий, имеющих одинаковую вероятность, зависит от заданных условий эксперимента и связи этих событий (равенство не гарантирует независимость или зависимость).

Два парохода должны подойти к одному и тому же причалу независимо друг от друга и равновозможно в течение суток. Определить вероятность того, что одному из них придется ожидать освобождения причала,
если время стоянки первого парохода — 1 час, а второго — 2 часа.

Пусть $x$ - время прибытия 1, тогда $x + 1$ - время отплытия 1, $[x,x+1]$ - время стоянки

Пусть $y$ - время прибытия 2, тогда $y + 2$ - время отплытия 2, $[y,y+2]$ - время стоянки

$x,y \in [0,24]$

Если 1 приехал раньше $x \geqslant y$, то ситуация, когда 2 придется ждать $y \leqslant x+1$.

Если 2 приехал раньше $x \leqslant y$, то ситуация, когда 1 придется ждать $x \leqslant y+2$.

Эта система - полоса между $x-2$ и $x+1$. При этом достоверное событие (= 1) это квадрат 24 на 24 на графике($x,y \in [0,24]$).

Тогда вероятность, что кто-то будет ждать: $P(A) = S(полосы)/S(квадрата)$

$S_кв = 24\cdot 24 = 576$

$S_пол = S_кв - S_вт - S_нт = 576 - \frac{1}{2}\cdot 23^2 - \frac{1}{2}\cdot 22^2 = 69.5$

$P(A) = 69.5/576 = 139/1152$

```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем область построения
fig, ax = plt.subplots(figsize=(5, 4))

# Определяем диапазон для x и y
x = np.linspace(0, 24, 400)
y = np.linspace(0, 24, 400)
X, Y = np.meshgrid(x, y)

# Определяем область между x-2 и x+1
condition = (Y >= X - 2) & (Y <= X + 1)

# Закрашиваем область между линиями
plt.contourf(X, Y, condition, levels=[0.5, 1.5], alpha=0.3, colors=['lightblue'])
plt.contour(X, Y, condition, levels=[0.5], colors='blue', linewidths=2)

# Рисуем граничные линии
plt.plot(x, x + 1, 'r-', label='y = x + 1', linewidth=2)
plt.plot(x, x - 2, 'g-', label='y = x - 2', linewidth=2)
plt.plot(x, x, 'k--', label='y = x', alpha=0.5, linewidth=1)

# Настраиваем график
plt.xlim(0, 24)
plt.ylim(0, 24)
plt.xlabel('Время прибытия 1 (x)', fontsize=12)
plt.ylabel('Время прибытия 2 (y)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()

# Устанавливаем одинаковый масштаб по осям
plt.gca().set_aspect('equal')

plt.tight_layout()
plt.show()
```

### 5

Самолет, по которому ведется стрельба, состоит из трех различных по уязвимости частей:
1) Кабина летчика и двигатель
2) Топливные баки
3) Планер

Для поражения самолета достаточно либо одного попадания в первую часть, либо двух попаданий во вторую, либо трех в третью. При попадании в самолет одного снаряда, снаряд с вероятностью $p1$ попадает в первую часть, с вероятностью $p2$ — во вторую, с вероятностью $p3$ — в третью. Попавшие снаряды распределяются по частям независимо друг от друга.

Известно, что в самолет попало $m$ снарядов. Найти условную вероятность $P(A|m)$ события $A$ – «Самолет поражен» – при $m = 1, 2, 3, 4$

Пусть $X_i$ - число попаданий в $i = 1, 2, 3$ часть самолета, $\sum X_i = m$

Тогда мы можем расписать вероятность $A$ как $P(A|m) = P(X_1 \geqslant 1) + P(X_2 \geqslant 2) + P(X_3 \geqslant 3)$ - это полиномиальное распределение, тк мы определеяем вероятность, что в серии испытаний каждое событие произойдет определенное количество раз. 

$$P(X_1 = x_1, X_2 = x_2, \ldots, X_k = x_k) = \frac{n!}{x_1! \cdot x_2! \cdot \ldots \cdot x_k!} \cdot p_1^{x_1} \cdot p_2^{x_2} \cdot \ldots \cdot p_k^{x_k}$$ 

Будем считать, что мы не попадаем в другие части самолета, кроме уязвимых, тогда $p_1 + p_2 +p_3 =1$



$m = 1$

Это случай, когда самолет сбили с первого раза и он возможет только в ситуации, когда с первого раза попали в 1 часть. 

$P(A|m = 1) = P(X_1 = 1) = p_1$

$m=2$

Чтобы не решать перебором рассмотрим вероятность ситуации, что самолет не повержен $\overline{A}$, вероятность которой можно записать как $P(\overline{A}|m) = P(X_1 = 0) + P(X_2 \leqslant 1) + P(X_3 \leqslant 2)$. 

Тогда мы можем найти искомую вероятность как $1-\overline{A}$.

При 2 попаданиях у нас есть 2 варианта, когда самолет останется не повержен:

1. $X_1 = 0, X_2 = 1, X_3 = 1$
2. $X_1 = 0, X_2 = 0, X_3 = 2$


Тогда вероятность $\overline{A}$ находим как $P(\overline{A}|m=2) = P(X_1 = 0, X_2 = 1, X_3 = 1) + P(X_1 = 0, X_2 = 0, X_3 = 2) = 2p_2p_3 +{p_3}^2$

$P(A|m = 2) = 1 - 2p_2p_3 - {p_3}^2$

$m=3$

При 3 попаданиях у нас есть только 1 вариант, когда самолет останется не повержен:

$$X_1 = 0, X_2 = 1, X_3 = 2$$

$P(\overline{A}|m=3) = P(X_1 = 0, X_2 = 1, X_3 = 2) = 3p_2{p_3}^2$

$P(A|m = 3) = 1 - 3p_2{p_3}^2$

$m=4$

При 4 попаданиях: 

$$\begin{cases}
X_1 + X_2 + X_3 = 4 \\
X_1 = 0 \\
X_2 \leqslant 1 \\
X_3 \leqslant 2
\end{cases}$$

$$\begin{cases}
X_1 + X_2 + X_3 = 4 \\
X_1 + X_2 + X_3 \leqslant 3 
\end{cases}$$

Получили противоречие, значит не существует ситуации, когда самолет останется не повержен $P(\overline{A}|m=4) = 0$

$P(A|m = 4) = 1$
